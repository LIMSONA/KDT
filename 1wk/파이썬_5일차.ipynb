````python
class Counter:
    
    def __init__(self):
        self.num = 10
    
    def up_one(self):
        self.num += 1
        
    def down_one(self):
        self.num -= 1
        
    def zero_reset(self):
        self.num = 0
        
    def value_info(self):
        print("í˜„ì¬ ê°’ì€ {}ì…ë‹ˆë‹¤!!!".format(self.num))
        
````

````python
now=Counter()

now.up_one()
now.down_one()
now.zero_reset()
now.value_info()
````

#### class method(static method) - classë¡œ í˜¸ì¶œ
- í´ë˜ìŠ¤ ë©”ì„œë“œì˜ ê²½ìš° ê°ì²´ê°€ ìƒì„±ë˜ê¸° ì „ì— ë¯¸ë¦¬ ìƒì„±ë¨ìœ¼ë¡œ ê°ì²´ë¥¼ í†µí•´ í˜¸ì¶œí•˜ì§€ ì•ŠëŠ”ë‹¤.
- **`@classmethod`**

````python
class Cal3:
    
    name="sona"
    
    def __init__(self,name):
        self.name = name
    
    def minus(self,a,b):
        print(self.name)
        return a-b
    
    @staticmethod
    def add(a,b):
        return a+b
    
    @classmethod
    def mul(cls,a,b):
        print("mul",cls.name)
        return a*b
````
