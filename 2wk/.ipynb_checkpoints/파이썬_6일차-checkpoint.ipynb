### **special method**
 - __ë¡œ ì‹œì‘ __ë¡œ ëë‚˜ëŠ” íŠ¹ìˆ˜ í•¨ìˆ˜ë‹¤.
 - í•´ë‹¹ ë©”ì˜ë“œë“¤ì„ êµ¬í˜„í•˜ë©´, ì»¤ìŠ¤í…€ ê°ì²´ì— ì—¬ëŸ¬ê°€ì§€ íŒŒì´ì¬ ë‚´ì¥ í•¨ìˆ˜ë‚˜ ì—°ì‚°ìë¥¼ ì ìš© ê°€ëŠ¥í•˜ë‹¤.
 - ì˜¤ë²„ë¼ì´ë”© ê°€ëŠ¥í•œ í•¨ìˆ˜ ëª©ë¡ì€ ì•„ë˜ ë§í¬ì—ì„œ ì°¸ì¡° 
   - https://docs.python.org/3/reference/datamodel.html
   - https://docs.python.org/3/reference/datamodel.html#object.__add__

````python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def __add__(self,p):
        print("__add__ êµ¬í˜„")
        new_x = self.x + p.x
        new_y = self.y + p.y
        return Point(new_x, new_y)
````

````python
p1 = Point(2,3)
p2 = Point(3,5)
````

````python
print(p1.x, p1.y)
print(p2.x, p2.y)
````

````python
p3 = p1 + p2

print(p3.x, p3.y)
````
